```python
#Importing the libraries

import PIL.Image
import pilgram
```

```python
img = PIL.Image.open("Hills.jpg") #Taken this photo on OnePlus 7T
# rotate_img = img.rotate(-90, expand=True) #The photo was rotating at counter clockwise direction
resized_img = img.resize((480,360)) #The resolution of the photo is 4000*3000 which is too large to display on the Notebook
display(resized_img)
```

```python
#Applying the lofi filter
#One of the oldest and the most famous instagram filters

img_lofi = pilgram.lofi(resized_img)#Saved photo will have the original resolution
display(img_lofi) 
```

```python
#Now we will see how multiple filters affect the photo
multifil_img = pilgram.hudson(pilgram.xpro2(pilgram.lofi(resized_img)))
display(multifil_img)
```

```python
#The image look too saturated because of using lofi filter. So, to reduce the saturation, the value would be set less than 1 to reduce it
multifil_img = pilgram.css.saturate(multifil_img, 0.8)
display(multifil_img)
```

```python
# Now the edited photo with original resloution will be stored
edit_img = pilgram.hudson(pilgram.xpro2(pilgram.lofi(img)))
edit_img = pilgram.css.saturate(edit_img, 0.8)
edit_img.save("Hills_pilgram_filters.jpg")
```
